```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/PIWorks_data/country_vaccination_stats.csv")
data = df.copy()
data
```

```python
missing_values = data[data['daily_vaccinations'].isna()]
print(missing_values)
```

```python
minimum_daily_vaccinations_by_country = data.groupby("country")["daily_vaccinations"].min()

def update_vaccinations(row):
    if pd.isnull(row["daily_vaccinations"]):
        if row["country"] in minimum_daily_vaccinations_by_country:
            return minimum_daily_vaccinations_by_country[row["country"]]
    else:
        return row["daily_vaccinations"]

data["daily_vaccinations"] = data.apply(update_vaccinations, axis=1)
data["daily_vaccinations"].fillna(0, inplace=True)
```

```python
median_daily_vaccinations_by_country = data.groupby('country')['daily_vaccinations'].median()
sorted_countries = median_daily_vaccinations_by_country.sort_values(ascending=False)
sorted_countries.head(3)
```
